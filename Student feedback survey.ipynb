```
import pandas as pd
df = pd.read_csv('student_feedback.csv')
df.head()
```

```
df.drop
```

```
df.drop(columns=['Unnamed: 0'], inplace=True)

df.head()
```

```
df.info()
df.isnull().sum()
```

```
df["Average_Score"] = df.mean(axis=1)

df[["Student ID", "Average_Score"]].head()
```

```
def satisfaction_category(score):
    if score >= 7:
        return "Positive"
    elif score >= 4:
        return "Neutral"
    else:
        return "Negative"

# Apply function
df["Satisfaction"] = df["Average_Score"].apply(satisfaction_category)

df[["Average_Score", "Satisfaction"]].head()
```

```
avg_scores = df.mean(numeric_only=True).sort_values(ascending=False)
avg_scores
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.barplot(
    x=avg_scores.index,
    y=avg_scores.values
)
plt.xticks(rotation=45, ha='right')
plt.title('Average Student Feedback Scores by Criteria')
plt.ylabel('Average Rating')
plt.xlabel('Feedback Criteria')
plt.show()
```

```
avg_ratings = df.drop(columns=["Student ID", "Average_Score", "Satisfaction"]).mean()

plt.figure()
avg_ratings.plot(kind="bar")
plt.title("Average Rating per Feedback Parameter")
plt.xlabel("Feedback Parameters")
plt.ylabel("Average Score")
plt.xticks(rotation=90)
plt.show()
```
